```python
#imports for mapping and graphing
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

#imports for ML
from sklearn.linear_model import LinearRegression
from sklearn import datasets
from sklearn.model_selection import train_test_split 

#load dataset
df = datasets.load_boston()

#select features from dataset
Y = df.target
X = df.data[:, [5,6]]

#split data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)

#build and train our model
regr = LinearRegression()
regr.fit(x_train, y_train)

#test the model
y_pred = regr.predict(x_test)

def plotter(fignum, elev, azim):
    fig = plt.figure(fignum, figsize=(3,4))
    plt.clf()
    ax = Axes3D(fig, elev=elev, azim=azim)
    ax.scatter(x_test[:, 0], x_test[:, 1], y_pred)
    ax.set_xlabel('Rooms')
    ax.set_ylabel('Age')
    ax.set_zlabel('Price')
    ax.set_title(f'View: {fignum+1}')

plotter(0, 20, 20)
plotter(1, 20, 90)
plotter(2, 90, 90)
print(f"Accuracy Score: {regr.score(x_test, y_test) * 100}")
plt.show()
```
