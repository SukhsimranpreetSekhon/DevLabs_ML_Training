```python
price1 = 0
price2 = 228
price3 = 1188

per_1000_m = 10
per_10g = 60
```

```python
#X Ranges for different versions of software

free_series_X = list(range(400, 91400, 1000))
free_series_X.insert(0, 0)
print(free_series_X)

prem_series_X = list(range(10000, 90000, 1000))
prem_series_X.insert(0,0)
print(prem_series_X)

ult_series_X = list(range(50000, 90000, 1000))
ult_series_X.insert(0,0)
print(ult_series_X)
```

```python
max_price = price1 + (len(free_series_X)) * 10

free_series_Y = list(range(0, max_price, 10))
print(free_series_Y)
```

```python
max_price = price2 + (len(prem_series_X)) * 10

prem_series_Y = list(range(price2, max_price, 10))
print(prem_series_Y)
```

```python
max_price = price3 + (len(ult_series_X)) * 10

ult_series_Y = list(range(price3, max_price, 10))
print(ult_series_Y)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot(free_series_X, free_series_Y, label="Community Edition")
ax.plot(prem_series_X, prem_series_Y, label="Premium Edition")
ax.plot(ult_series_X, ult_series_Y, label="Ultimate Edition")
ax.set_xlabel('CI/CD Minutes Used')
ax.set_ylabel('Price (USD)')
ax.legend()

plt.show()
```
